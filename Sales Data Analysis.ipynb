# Outlier Analysis

```python
col=['Age','Occupation','Stay_In_Current_City_Years', 'Marital_Status', 'Product_Category_1','Product_Category_2']
```

```python
for c in col:
  percentile25 = df1[c].quantile(0.25)
  percentile75 = df1[c].quantile(0.75)
  IQR = percentile75 - percentile25
  Upperlimit = percentile75 + 1.5*IQR
  Lowerlimit = percentile25 - 1.5*IQR
  df1= df1[df1[c]<=Upperlimit]
  df1 = df1[df1[c]>=Lowerlimit]
  plt.figure()
  sns.boxplot(y=c, data =df1)
```

# Onehot Encoding

```python
df= pd.get_dummies(df,columns=["City_Category"],drop_first=True)
```

```python
df.columns
```

```python
x=df[['User_ID', 'Gender', 'Age', 'Occupation', 'Stay_In_Current_City_Years',
       'Marital_Status', 'Product_Category_1', 'Product_Category_2',
       'Product_Category_3','City_Category_B', 'City_Category_C']].values
```

```python
y=df["Purchase"].values
```

# Linear Regression

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.2, random_state =25)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test =sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)

y_pred1 = reg.predict(x_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
r2_score(y_train, y_pred1)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_train,y_pred1)
```

```python
np.sqrt(mean_squared_error(y_train,y_pred1))
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dc=DecisionTreeRegressor()
```

```python
dc.fit(x_train,y_train)
```

```python
y_pred=dc.predict(x_test)
print("R2 Score for Testing data is",round((r2_score(y_test,y_pred)),4)*100)
```

```python
y_pred=dc.predict(x_train)
print("R2 Score for Testing data is",round((r2_score(y_train,y_pred)),4)*100)
```
