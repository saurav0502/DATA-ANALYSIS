#### Check the outliers for various Thresholds (1.5 x IQR , 2 X IQR and so on) for Numeric Columns

```python
df.columns
```

```python
col = ['number', 'age (years)', 'ethnicity',
       'mean diastolic bp (mmhg)', 'mean systolic bp (mmhg)',
       'central armellini fat (mm)','pregnancies (number)', 'first fasting glucose (mg/dl)',
       'bmi pregestational (kg/m)', 'child birth weight (g)']
```

```python
for c in col:
  percentile25 = df[c].quantile(0.25)
  percentile75 = df[c].quantile(0.75)
  IQR = percentile75 - percentile25
  Upperlimit = percentile75 + 1.5*IQR
  Lowerlimit = percentile25 - 1.5*IQR
  df= df[df[c]<=Upperlimit]
  df = df[df[c]>=Lowerlimit]
  plt.figure()
  sns.boxplot(y=c, data =df)
```

```python
for c in col:
  percentile25 = df[c].quantile(0.25)
  percentile75 = df[c].quantile(0.75)
  IQR = percentile75 - percentile25
  Upperlimit = percentile75 + 2*IQR
  Lowerlimit = percentile25 - 2*IQR
  df= df[df[c]<=Upperlimit]
  df = df[df[c]>=Lowerlimit]
  plt.figure()
  sns.boxplot(y=c, data =df)
```

#### Classify the dataset in the various reasonable clusters (Use K-Means), and explain the feature selection for clusters

```python
from sklearn.cluster import KMeans
```

```python
inertia=[]
for i in range(1,15):
    kmeans=KMeans(i)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
```

```python
sns.lineplot(x=range(1,15),y=inertia,marker="o")
```

```python
kmeans=KMeans(6)
```

```python
kmeans.fit(df)
```

```python
kmeans.predict(df)
```

```python
df["clusters"]=kmeans.predict(df)
```
